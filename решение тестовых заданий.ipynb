```python
import pandas as pd

df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные об аудитории')
df['date'] = pd.to_datetime(df['date'])

cohort_nov1 = set(df[df['date'] == '2023-11-01']['user_id'])

active_nov2 = set(df[df['date'] == '2023-11-02']['user_id'])

returned_users = cohort_nov1.intersection(active_nov2)

retention_d1 = len(returned_users) / len(cohort_nov1) * 100

print(f"Количество новых пользователей (01.11): {len(cohort_nov1)}")
print(f"Вернулись на следующий день (02.11): {len(returned_users)}")
print(f"Retention 1-го дня: {retention_d1:.2f}%")
```

```python
import pandas as pd

df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные об аудитории')

total_users = df['user_id'].nunique()

viewers = df[df['view_adverts'] > 0]['user_id'].nunique()

conversion = (viewers / total_users) * 100

print(f"Всего пользователей: {total_users}")
print(f"Пользователей с просмотрами: {viewers}")
print(f"Конверсия в просмотр: {conversion:.2f}%")
```

```python
import pandas as pd
from scipy import stats

df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные АБ тестов')

# 2. Проходим циклом по каждому из 3-х экспериментов
for exp in sorted(df['experiment_num'].unique()):
    data = df[df['experiment_num'] == exp]
    
    test_rev = data[data['experiment_group'] == 'test']['revenue']
    ctrl_rev = data[data['experiment_group'] == 'control']['revenue']
    
    # Считаем ARPU
    arpu_test = test_rev.mean()
    arpu_ctrl = ctrl_rev.mean()
    
    # Проводим t-тест (используем тест Уэлча, так как размеры групп и дисперсии могут отличаться)
    t_stat, p_val = stats.ttest_ind(test_rev, ctrl_rev, equal_var=False)
    
    print(f"--- Эксперимент №{exp} ---")
    print(f"ARPU Test: {arpu_test:.2f} | ARPU Control: {arpu_ctrl:.2f}")
    print(f"Прирост ARPU: {((arpu_test/arpu_ctrl)-1)*100:.2f}%")
    print(f"p-value: {p_val:.4f}")
    
    if p_val < 0.05:
        print("Результат: СТАТИСТИЧЕСКИ ЗНАЧИМ (разница есть)")
    else:
        print("Результат: НЕЗНАЧИМ (разница может быть случайной)")
    print("\n")
```

```python
import pandas as pd

df_listers = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Листеры')

total_revenue = df_listers['revenue'].sum()

unique_users = df_listers['user_id'].nunique()

arpu_listers = total_revenue / unique_users

print(f"Общий доход: {total_revenue}")
print(f"Уникальных пользователей (листеров): {unique_users}")
print(f"Средний доход на пользователя (ARPU): {arpu_listers:.2f}")
```
